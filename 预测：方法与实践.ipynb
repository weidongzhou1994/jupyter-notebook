# 时间序列图形

## ts对象 

&emsp;&emsp;在 R 语言中，可以使用 ts 函数创建时间序列的对象，ts 函数的基本语法如下：

```r
y <- ts(c(123, 39, 78, 52,110), start = 2008, frequency = 1 
        # 第一个参数是观测值，第二个参数四起始年份，第三个参数是频率
y
```

&emsp;&emsp;对于第二个参数，还可以通过使用 c(年， 月) 指定起始年月；频率可供选择的值如下：

|数据类型|频率|
|--|--|
|年度|1|
|季度|4|
|月度|12|
|周|52|


&emsp;&emsp;更为具体的用法可以查看帮助：

```r
?ts
```

## 时间图 

```r
library(fpp2)
library(ggpomological)

autoplot(melsyd[,"Economy.Class"]) +
    ggtitle("墨尔本-悉尼经济舱乘客客流量") +
    xlab("年份") +
    ylab("千") +
    theme_pomological(base_family = windowsFont("kaiti"), base_size = 15) +
    theme(plot.title = element_text(hjust = 0.5))
```

&emsp;&emsp;观看时间图通常可以简单地判断时间序列数据的趋势、季节性和周期性。趋势是指时间序列数据长期增长或下降；季节性是指数据存在频率不变且与长度固定的时间段有关的波动；周期性是指数据存在不固定频率的上升或下降。

![](https://i.loli.net/2020/06/17/QiNojzUBGSXOtWp.png)

&emsp;&emsp;明显，英国新建房屋销售额存在强烈的年度季节性以及周期为 6~10 年的周期性。

![](https://i.loli.net/2020/06/17/VtnkjyNcrHuXpGA.png)

&emsp;&emsp;明显，澳大利亚季度电力产值存在强烈的年度季节性以及向上增长的趋势。

&emsp;&emsp;值得注意的是：只有存在趋势、季节性或周期性的时间序列数据才有可能进行预测，对于没有良好形态特性的随机波动，预测效果会大打折扣，如：

![](https://i.loli.net/2020/06/17/vasNjXCDlKhq5td.png)

## 季节图

&emsp;&emsp;为了更好地判断时间序列数据是否具有季节性特征，可以绘制季节图进行查看：

```r
ggseasonplot(a10, year.labels=TRUE, year.labels.left=TRUE) +
  xlab("月份")+
  ylab("百万（美元）") +
  ggtitle("季节图：降糖药物销量")+
  theme_pomological(base_family = windowsFont("kaiti"), base_size = 15) +
  theme(plot.title = element_text(hjust = 0.5))
```

&emsp;&emsp;也可以通过设置参数 polar=TRUE 转化为极坐标： 

```r
ggseasonplot(a10, polar=TRUE) +
  xlab("月份")+
  ylab("百万（美元）") +
  ggtitle("季节图：降糖药物销量")+
  theme_pomological(base_family = windowsFont("kaiti"), base_size = 15) +
  theme(plot.title = element_text(hjust = 0.5))
```

&emsp;&emsp;明显，从季节图中可以看出：12月~1月的季节性十分明显，从极坐标图中可以进一步看出这种季节性。进一步，我们可以通过 ggsubseriesplot() 着重描绘在相同月份数据的同比变化情况：

```r
ggsubseriesplot(a10) +
  xlab("月份")+
  ylab("百万（每月）") +
  ggtitle("子序列季节图：降糖药物销量") +
  theme_pomological(base_family = windowsFont("kaiti"), base_size = 15) +
  theme(plot.title = element_text(hjust = 0.5))
```

## 散点图

&emsp;&emsp;当我们研究两个变量之间的关系时，可以在同一张图上绘制两个时间序列的散点图，比如用电量和温度之间的关系：

```r
qplot(Temperature, Demand, data=as.data.frame(elecdemand)) +
  ylab("用电量 (千兆瓦)") + xlab("温度 (摄氏度)")+
  theme_pomological(base_family = windowsFont("kaiti"), base_size = 15) +
  theme(plot.title = element_text(hjust = 0.5))
```

&emsp;&emsp;上述散点图简单描绘了两个变量之间的关系，一个定量的指标是相关系数：

$$
r = \frac{\sum (x_{t} - \bar{x})(y_{t}-\bar{y})}{\sqrt{\sum(x_{t}-\bar{x})^2}\sqrt{\sum(y_{t}-\bar{y})^2}}.
$$

## 滞后图

```r
beer2 <- window(ausbeer, start=1992) # window()函数可以截取时间序列的一部分
gglagplot(beer2) + # 默认滞后阶数为9
    theme_pomological(base_family = windowsFont("kaiti"), base_size = 15) 
```

## 自相关

&emsp;&emsp;对于时间序列滞后值之间的线性关系，可以使用自相关系数进行衡量：

$$
r_{k} = \frac{\sum\limits_{t=k+1}^T (y_{t}-\bar{y})(y_{t-k}-\bar{y})}
 {\sum\limits_{t=1}^T (y_{t}-\bar{y})^2}
$$

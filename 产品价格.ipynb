```python
import requests
```

```python
url = 'http://www.stats.gov.cn/tjsj/zxfb/202008/t20200804_1780431.html'

html = requests.get(url)
text = html.content.decode("utf8")
```

```python
from lxml import etree

html = etree.HTML(text)
```

```python
data_1 = []
for i in range(1,60):
    address = '//table/tbody/tr[' + str(i) + ']/td[1]/p/span'
    data1 = html.xpath(address)
    data = ''
    for j in range(0, len(data1)-1):
        data = data + str(data1[j].text)
    data_1.append(data)
data_1
```

```python
data_2 = []
for i in range(1,60):
    address = '//table/tbody/tr[' + str(i) + ']/td[2]/p/span'
    data1 = html.xpath(address)
    data = data1[0].text
    data_2.append(data)
data_2
```

```python
data_3 = []
for i in range(1,60):
    address = '//table/tbody/tr[' + str(i) + ']/td[3]/p/span'
    data1 = html.xpath(address)
    data = data1[0].text
    data_3.append(data)
data_3
```

```python
data_4 = []
for i in range(1,60):
    address = '//table/tbody/tr[' + str(i) + ']/td[4]/p/span'
    data1 = html.xpath(address)
    data = data1[0].text
    data_4.append(data)
    
data_4
```

```python
data_5 = []
for i in range(1,60):
    address = '//table/tbody/tr[' + str(i) + ']/td[5]/p/span'
    data1 = html.xpath(address)
    try:
        data = data1[0].text
        data_5.append(data)
    except:
        data = ''
        data_5.append(data)
data_5
```

```python
import pandas as pd

data = pd.DataFrame({
    '产品名称': data_1,
    '单位': data_2,
    '本期价格（元）': data_3,
    '比上期价格涨跌（元）': data_4,
    '涨跌幅（%）': data_5
})
data = data[-data['产品名称'].isin([''])]
data
```
